```python
polinomial = lambda x : 3*(x**2) - 2*x + 1

exponencial = lambda x : x**2

senoidal = lambda x : cos(x) + sin(x)

def trapezoidal_integral(f, a, b, n=1):
    h = (b - a) / n  # Tamanho do intervalo
    integral = 0
    valores = []

    # Somar os valores dos trapézios
    for i in range(n):
        x0 = a + i * h
        x1 = a + (i + 1) * h
        
        #print('intervalo [{},{}]: '.format(x0, x1),(f(x0) + f(x1)) * h / 2)
        valores.append('[{}, {}]: '.format(x0, x1) + str((f(x0) + f(x1)) * h / 2))
        integral += (f(x0) + f(x1)) * h / 2
    
    #print('total: {}\n'.format(integral))
    return valores, integral

def curve_fitting(x, y, degree):
    coefficients = np.polyfit(x, y, degree)  # Realiza o ajuste polinomial
    fitted_curve = np.poly1d(coefficients)   # Gera a curva ajustada
    
    return fitted_curve

def xds(func, pontos=[-2, -1, 0, 1, 2]):
    integrais, total = trapezoidal_integral(func, -2, 2, 4) #INTEGRAL NUMERICA DA FUNÇÃO

    dataarea = pd.DataFrame({
            'x':np.linspace(-2, 2, num=100)
        })
    dataarea['y'] = [func(i) for i in dataarea.x]
    
    dataarea['interval'] = pd.cut(dataarea.x, pontos, include_lowest=True, labels=integrais, ordered=False).astype('str')
    print(integrais)
    return dataarea, total

def curve_fitting(x, y, degree):
    coefficients = np.polyfit(x, y, degree)  # Realiza o ajuste polinomial
    fitted_curve = np.poly1d(coefficients)   # Gera a curva ajustada
    
    return fitted_curve

def calcintegral(func, x):
    integrais = []
    legenda = []

    for i in x:
        auxintegrais, total = trapezoidal_integral(func, i, i-0.001, 1)
        integrais.append(total)
        legenda.append(auxintegrais)


    dataarea = pd.DataFrame({
            'x':np.linspace(-2, 2, num=100)
        })
    dataarea['y'] = [func(i) for i in dataarea.x]

    dataintegral = pd.DataFrame({
        'x':x,
        'integrais':integrais,
        'legenda':legenda
    })
    dataintegral['y'] = [func(i) for i in x]
    return dataintegral, dataarea
```

```python
x = [-2, -1, 0, 1, 2]

dataintegral, dataarea = calcintegral(polinomial, x)
dataintegral

grafico_pontos = alt.Chart(dataintegral).mark_bar().encode(
    x='x',
    y='y',
    tooltip = 'legenda'
)

grafico_pontos


#print(integrais)
```
